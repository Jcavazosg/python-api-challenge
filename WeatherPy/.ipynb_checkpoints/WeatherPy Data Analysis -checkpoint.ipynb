# Observable Trends for WeatherPy Analysis: 

1. Temperature peaks at around 20 degress latitude, which is different from what most people would expect, that the highest temperatures would be observed at the equator. Also, as expected decreasing temperatures can be observed as distance from the equator increases. 

2. Cloudiness does not seem to show any definitive corralation to latitude, data points observed are too scattered over the lattitude range. On the other hand the humidity data gathered shows somewhat of a trend to be higher at higher latitudes, but from the data observed there is still a lot of variance. A lager data set would help us identify if there is a definite corralation on humidity and latitude. 

3. There is no claer strong relationship between latitude and wind speed. Wind speed tends to generally be betweeen 0 and 15 mph regardless of latitude. 


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from api_keys import api_key


import api_keys

from citipy import citipy

output_data_file = "output_data/cities.csv"

lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List 

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
api_key = api_keys.api_key
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
```

```python
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

print(f"Starting Data Retrival")
print(f"-----------------------------------")

for city in cities:  
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} | {city_record}")
        print(f"{url}&q={city}")
        
        record = record + 1
            
    except:
        print("City not found. Skipping City -------")
    continue
```

```python
weatherpydata_dict = {
    "City": city_name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}

weather_data = pd.DataFrame(weatherpydata_dict)
weather_data.count()
```

```python
weather_data.to_csv('Output_CSV/weather_data.csv')
weather_data.head()
```

## Latitude vs. Temperature Plot

```python
plt.scatter(weather_data["Lat"], weather_data["Max Temp"], marker="o", s=10)
plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

plt.savefig("Output_Plots/Max_Temp_vs_Latitude.png")
plt.show()
```
